### Initial Code Chunk for accessing API endpoints

Using this code chunk multiple times and storing the results in different variables for cleaning and implementation

```python
# endpoint for retrieving different portions of the API
# ONLY RUN IF HAVE TO, uses one of 100 daily requests allowed with free account access
url = "https://v3.football.api-sports.io/countries"

payload = {}
headers = {
  'x-rapidapi-key': '01914987bf14c6a4b868a79dfe489f07',
  'x-rapidapi-host': 'v3.football.api-sports.io'
}

response = requests.request("GET", url, headers = headers, data = payload)

print(response.json())
```

#### Processing Timezones (completed, information exported to api_data/regions.csv)

```python
# first pull timezone response from endpoint code chunk
timezones_response = response.json()
```

```python
# extracting response from code 
timezones = timezones_response['response']
timezones
```

```python
# convert timezones (list) to a dictionary
# keys will be the region (first part of string before the slash)
# values will be a list of the cities/subregions within that region
# value lists should naturally be in alphabetical order because of nature
# timezone response 
timezone_dict = {}
for entry in timezones:
    region, city = entry.split('/', 1)
    timezone_dict.setdefault(region, []).append(city)
```

```python
# creating timezone dataframe to store in csv file
timezone_df = pd.DataFrame([tz.split('/', 1) for tz in timezones], columns=['Region', 'City'])
```

```python
# export to csv file in main directory
timezone_df.to_csv("api_data/regions.csv", index = False)
```

#### Processing Countries (completed, information exported to api_data/countries.csv)

```python
# first save countries response in separate variable
countries_response = response.json()
countries_response
```

```python
# pull response key from json format
countries = countries_response['response']
countries
```

```python
# convert list of dictionaries to pd.DataFrame
countries_df = pd.DataFrame(data = countries)
```

```python
# replace dashes and 'And's in country names with spaces and ampersands
countries_df['name'] = countries_df['name'].str.replace(pat = '-And-', repl = ' and ')
countries_df['name'] = countries_df['name'].str.replace(pat = '-', repl = ' ')

# remove row for World, can remove this code if the World 'country' is needed in the future
# also reindex the dataframe so numbers are consistent
countries_df.drop(167)
countries_df = countries_df.reset_index(drop = True)
```

```python
# save results to csv file for later use (jic)
countries_df.to_csv("api_data/countries.csv", index = False)
```
